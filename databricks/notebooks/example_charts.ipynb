## Charts in Databricks

Databricks has default charting that is perfect for a quick data science workflow.  In addition they allow you to access R and Python's plotting libraries as usual. This notebook was developed from [their materials](https://docs.databricks.com/notebooks/visualizations/index.html)

### Loading data

You can read the __example_data__ notebook to see details on loading data. This notebook defaults to Python so we need `%r` for the R chunks.

```
# Python load
mpg_df = spark.read.csv("/databricks-datasets/Rdatasets/data-001/csv/ggplot2/mpg.csv", header = "true", inferSchema = "true")
mpg = mpg_df.toPandas()
```

```
%r
# R load
library(SparkR)
mpg_df <- read.df("/databricks-datasets/Rdatasets/data-001/csv/ggplot2/mpg.csv", "csv", header = "true", inferSchema = "true", na.strings = "NA")
mpg <- SparkR::collect(mpg_df)
```

## Examples using Python

```
# %pip install altair
# https://docs.databricks.com/libraries/notebooks-python-libraries.html#pip-install
```

```
display(mpg_df)
```

```
# https://altair-viz.github.io/altair-tutorial/notebooks/06-Selections.html
import altair as alt

chart = alt.Chart(mpg) \
  .encode(
    alt.X('cty', scale=alt.Scale(zero=False)),
    alt.Y('hwy', scale=alt.Scale(zero=False)),
    tooltip = 'model',
    color = 'manufacturer'
  ) \
  .mark_circle() \
  .properties(width = 500, height = 300) \
  .interactive()

chart
```
